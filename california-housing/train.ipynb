```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from sklearn.datasets import fetch_openml
```

```python
data = fetch_openml(name="california_housing",as_frame=True)
```

```python
housing = pd.DataFrame(data.data,columns=data.feature_names)
```

```python
housing.head()
```

```python
housing.to_csv("california_housing.csv",index=False)
print("Saved the California housing dataset to california_housing.csv")
```

```python
from google.colab import files
files.download("california_housing.csv")
```

```python
housing.describe()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=housing, x='MedInc', y='AveRooms', alpha=0.35, s=20)
plt.title('Scatter Plot: MedInc vs AveRooms')
plt.xlabel('Median Income')
plt.ylabel('Average Rooms')
plt.grid(True, linestyle='--', alpha  =0.3)
plt.show()
```

```python
plt.scatter(housing['MedInc'],data.target,alpha=0.2,s=20)
```

```python
housing["price"] = data.target
housing.head(2)
```

```python
plt.scatter(housing['MedInc'],housing['price']*1000, alpha=0.2, s=20)
```
